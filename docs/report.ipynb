# **2. Task 1 – Mountain Named Entity Recognition**

## Objective

The objective of this task was to develop a Named Entity Recognition (NER) system capable of identifying mountain names. The solution was expected to recognize mountain entities within sentences while distinguishing them from the surrounding context.

The task involved preparing a suitable training dataset, selecting an appropriate transformer-based model, fine-tuning it for the custom entity type, and evaluating its performance on unseen examples.

## Dataset

Since no ready-to-use dataset specifically designed for mountain name recognition was available, a custom dataset was created for this project.

First, a collection of real mountain names was obtained from the Open Peaks dataset. These names served as the basis for generating annotated training examples.

Using the collected mountain names, synthetic sentences were automatically generated with the assistance of LLM. Each sentence contained one or more mountain names labeled using the BIO tagging scheme. The generated dataset was then converted into a format suitable for transformer-based token classification models.

This approach provided a diverse and well-structured training dataset while ensuring that all mountain names corresponded to real geographical entities.

## Model Selection

For this task, the pretrained **bert-base-uncased** model was selected as the foundation for the Named Entity Recognition system.

BERT was chosen because it has demonstrated strong performance on a wide range of Natural Language Processing tasks, including token classification and named entity recognition. Using a pretrained transformer allows the model to leverage contextual language representations learned from large-scale text corpora while requiring only task-specific fine-tuning.

The model was fine-tuned for token classification using the custom mountain NER dataset prepared for this project. 

## Training

The model was fine-tuned using the custom mountain NER dataset prepared during the data generation stage. Before training, the dataset was tokenized with the corresponding BERT tokenizer, and the BIO entity labels were aligned with the generated tokens.
The Hugging Face Transformers library was used to simplify the fine-tuning process. 
After fine-tuning, the trained model and tokenizer were saved and later used for inference on previously unseen sentences containing mountain names.

## Results

After fine-tuning, the model was able to identify mountain names and correctly distinguish them from other words in the text.

The trained model recognized both single-word and multi-word mountain names while preserving the BIO labeling scheme. An inference pipeline was implemented to demonstrate predictions on custom input sentences, and a demonstration notebook was included in the repository for interactive testing.


### Evaluation Metrics

The model was evaluated on the validation dataset after the final training epoch.

- **Precision:** 1.000
- **Recall:** 1.000
- **F1-score:** 1.000
- **Accuracy:** 1.000
- **Validation Loss:** 0.000050

## Limitations

Although the model achieved excellent results on the validation dataset, several limitations should be considered.

The training dataset was generated automatically using real mountain names combined with synthetic sentence templates. To improve robustness, the dataset also included negative examples containing people, cities, companies, organizations, technologies, and other non-mountain entities. However, the generated sentences cannot fully capture the diversity and complexity of real-world text.

The current model is designed specifically for mountain name recognition and has not been trained to identify other types of named entities.

Finally, the model was evaluated only on the held-out validation dataset created for this project. Additional evaluation on real-world documents would provide a more comprehensive assessment of its generalization ability.

## Future Improvements

Several enhancements could further improve the current solution.

The training dataset could be expanded with real-world documents containing mountain names, such as Wikipedia articles, travel guides, or geographical descriptions. This would expose the model to more diverse writing styles and sentence structures while preserving the same mountain recognition task.

Support for additional geographical entities, including rivers, lakes, cities, and countries, could also be added by extending the annotation scheme and retraining the model.

Finally, the model could be evaluated on real-world documents, such as Wikipedia articles or travel guides containing mountain names, to better assess how well it generalizes beyond the synthetic validation dataset used in this project.

# **3. Task 2 – Sentinel Image Matching**

## Objective

The objective of this task was to identify corresponding regions between two Sentinel-2 satellite images. The solution was expected to find matching image tiles by comparing local visual features and selecting the pair with the highest similarity.

Rather than training a model from scratch, the task focused on building a robust image matching pipeline using a pretrained deep learning model together with appropriate image preprocessing and tile-based comparison.

## Dataset

The task was performed using two Sentinel-2 satellite images provided as part of the assignment. The images were stored in the standard `.SAFE` directory structure and contained True Color Image (TCI) data.

Due to the high resolution of the original images, preprocessing was required before matching. Black border regions were removed, and each image was divided into fixed-size tiles. Only valid image tiles containing meaningful visual information were considered during the matching process, reducing unnecessary computations and improving the efficiency of the pipeline.


## Methodology

A tile-based image matching pipeline was implemented to identify corresponding regions between the two satellite images.

First, the Sentinel-2 spectral bands B04 (red), B03 (green), and B02 (blue) were loaded using the `rasterio` library. Contrast stretching was applied to each band before combining them into an RGB image.

Since Sentinel-2 images have a high spatial resolution, each image was divided into fixed-size tiles to make the matching process more efficient.

Only tiles with the same spatial position in both images were compared. This significantly reduced the computational cost while preserving the expected geographical correspondence between the images.

Feature matching was then performed using the pretrained LoFTR model from the Kornia library. Tile pairs were ranked primarily by the number of detected feature correspondences, while the average confidence score produced by LoFTR was used as a secondary criterion to resolve ties.

Finally, the tile pair with the highest ranking score was selected as the best match.

## Results

The implemented pipeline successfully identified the most similar tile pair between the two Sentinel-2 images. By combining tile-based preprocessing with the pretrained LoFTR model, the solution was able to detect corresponding image regions despite seasonal changes in appearance.

The modular design of the pipeline separates image loading, preprocessing, feature matching, and result selection into independent components, making the implementation easier to understand, maintain, and extend.

The complete inference workflow and qualitative matching results are demonstrated in the accompanying demo.ipynb notebook.

## Limitations

The current implementation assumes that the input images are approximately aligned, since only tiles with the same spatial position are compared. If the images were significantly shifted or rotated, additional image registration or a more exhaustive search strategy would be required.

The pipeline uses only the visible Sentinel-2 spectral bands (B04, B03, and B02) to construct RGB images. Additional spectral bands, such as the near-infrared or short-wave infrared channels, were not incorporated and could provide complementary information for more challenging matching scenarios.

Finally, tile ranking is primarily based on the number of detected feature correspondences, with the average confidence score produced by LoFTR used only as a secondary ranking criterion. More advanced scoring strategies could further improve the robustness of the matching process.

## Future Improvements

Several enhancements could further improve the current pipeline.

The matching process could be extended to compare neighboring tiles in addition to tiles with identical spatial positions. This would increase robustness when satellite images are not perfectly aligned.

Additional Sentinel-2 spectral bands, such as the near-infrared or short-wave infrared channels, could be incorporated to provide complementary information and improve matching performance under more challenging conditions.

The pipeline could also be evaluated with alternative deep learning-based feature matching models, such as SuperPoint + LightGlue or SuperGlue. Comparing multiple approaches could help identify the most suitable model for different satellite imagery conditions.
