# Mountain Named Entity Recognition Demo

This notebook demonstrates a fine-tuned BERT model for recognizing mountain names in English text.

The project includes:

- automatic collection of real mountain names;
- synthetic dataset generation with BIO labels;
- positive and negative training examples;
- fine-tuning of `bert-base-uncased` for token classification;
- inference on previously unseen sentences.


## Pipeline Overview

The complete workflow consists of five stages:

1. Download mountain names from the Open Peaks dataset.
2. Generate a synthetic NER dataset using BIO tagging.
3. Fine-tune a BERT model for token classification.
4. Save the trained model.
5. Run inference on unseen text.

```python
from pathlib import Path
import sys

PROJECT_ROOT = Path.cwd().parent
if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))
```

```python
import pandas as pd
from IPython.display import Markdown, display
from inference import predict_mountains
```

```python
examples = [
    "Although the weather forecast looked terrible, our group finally reached Mount Elbrus before sunset.",
    "OpenAI released a new multimodal model yesterday.",
    "John visited Paris during the summer.",
    "The helicopter circled above Hinterer Seelenkogel before the rescue operation began.",
    "Artificial intelligence is transforming modern healthcare.",
    "During our expedition we climbed Mount Elbrus first and visited Mont Blanc two weeks later.",
]
for text in examples:

    mountains = predict_mountains(text)

    display(Markdown(f"### Input\n\n{text}"))

    if mountains:
        display(Markdown(
            "### Detected mountain(s)\n\n"
            + "\n".join(f"- **{m.title()}**" for m in mountains)
        ))
    else:
        display(Markdown(
            "### Detected mountain(s)\n\n"
            "*No mountain names found.*"
        ))

    print("-" * 80)
```

## Conclusion

A BERT-based Named Entity Recognition model was successfully fine-tuned to identify mountain names in English text.

The inclusion of automatically generated negative examples improved the model's robustness by reducing false positive predictions on non-mountain entities such as cities, companies, organizations, and technologies.

The project provides both a command-line interface (`inference.py`) and an interactive Jupyter notebook (`demo.ipynb`) for convenient inference and demonstration of the trained model.
